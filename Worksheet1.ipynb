#3 Image Compression and Decompression using PCA.
In this exercise, build a PCA from scratch using explained variance method for image compression task.
You are expected to compute the necessary matrices from the scratch. Dataset: Use image of your
choice.
1. Load and Prepare Data:
• Fetch an image of you choice.{If colour convert to grayscale}
• Center the dataset - Standaridze the Data.
• Calculate the covaraince matrix of the Standaridze data.#

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def load_and_prepare_image(image_path):
    """Loads an image, converts it to grayscale, and prepares it for PCA."""
    img = Image.open(image_path).convert('L')  # Load the image and convert to grayscale ('L' mode)
    img_array = np.array(img)  # Convert the image into a numpy array
    return img_array  # Return the image data as a numpy array

def center_and_standardize(data):
    """Centers and standardizes the data."""
    mean = np.mean(data, axis=0)  # Compute the mean of the data across columns
    centered_data = data - mean  # Subtract the mean from the data (centering)
    std = np.std(centered_data, axis=0)  # Compute the standard deviation of the centered data
    standardized_data = centered_data / (std + 1e-8)  # Divide by the standard deviation (standardizing), add epsilon to avoid division by zero
    return standardized_data  # Return the standardized data

def calculate_covariance_matrix(data):
    """Calculates the covariance matrix."""
    return np.cov(data.T)  # Compute the covariance matrix (note: data is transposed for proper calculation)

# Example usage
image_path = '/content/drive/MyDrive/Artifical Intelligence and machine learning/bike.png'  # Path to the image file
image_data = load_and_prepare_image(image_path)  # Load and prepare the image
rows, cols = image_data.shape  # Get the shape of the image (rows and columns)
flattened_data = image_data.flatten().reshape(rows, cols)  # Flatten the image data into a 1D array and reshape it
standardized_data = center_and_standardize(flattened_data)  # Standardize the flattened data
covariance_matrix = calculate_covariance_matrix(standardized_data)  # Calculate the covariance matrix from the standardized data
```

2. Eigen Decomposition and Identifying Principal Components:

• Compute Eigen Values and Eigen Vectors.

• Sort the eigenvalues in descending order and choose the top k eigenvectors corresponding to the highest eigenvalues.

• Identify the Principal Components with the help of cumulative Sum plot.

```python
def eigen_decomposition(covariance_matrix):
    """Performs eigen decomposition."""
    eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)  # Calculate eigenvalues and eigenvectors of the covariance matrix
    return eigenvalues, eigenvectors  # Return the eigenvalues and eigenvectors

def select_principal_components(eigenvalues, eigenvectors, explained_variance_threshold=0.95):
    """Selects principal components based on explained variance."""
    sorted_indices = np.argsort(eigenvalues)[::-1]  # Sort the eigenvalues in descending order
    sorted_eigenvalues = eigenvalues[sorted_indices]  # Sort the eigenvalues based on the sorted indices
    sorted_eigenvectors = eigenvectors[:, sorted_indices]  # Sort the eigenvectors based on the sorted indices

    total_variance = np.sum(sorted_eigenvalues)  # Calculate the total variance (sum of eigenvalues)
    explained_variance_ratio = sorted_eigenvalues / total_variance  # Calculate the explained variance ratio for each eigenvalue
    cumulative_variance = np.cumsum(explained_variance_ratio)  # Compute the cumulative explained variance

    k = np.argmax(cumulative_variance >= explained_variance_threshold) + 1  # Find the number of components that explain at least the threshold variance
    selected_eigenvectors = sorted_eigenvectors[:, :k]  # Select the top k eigenvectors (principal components)

    return selected_eigenvectors, cumulative_variance  # Return the selected eigenvectors and cumulative variance

# Example usage
eigenvalues, eigenvectors = eigen_decomposition(covariance_matrix)  # Perform eigen decomposition on the covariance matrix
selected_pcs, cumulative_variance = select_principal_components(eigenvalues, eigenvectors)  # Select principal components based on explained variance

# Plot cumulative variance
plt.plot(cumulative_variance)  # Plot the cumulative explained variance as a function of the number of principal components
plt.xlabel('Number of Principal Components')  # Label for x-axis
plt.ylabel('Cumulative Explained Variance')  # Label for y-axis
plt.title('Cumulative Explained Variance Plot')  # Title of the plot
plt.show()  # Display the plot

print(f"Number of selected principal components: {selected_pcs.shape[1]}")  # Print the number of selected principal components
```

3. Reconstruction and Experiment:

• Reconstruction: Transform the original data by multiplying it with the selected eigenvec-
tors(PCs) to obtain a lower-dimensional representation.

• Experiments: Pick Four different combination of principal components with various ex-
plained variance value and compare the result.

• Display the Results and Evaluate.

```python
import matplotlib.pyplot as plt
import numpy as np

def reconstruct_image(data, selected_eigenvectors, mean, std):
    """Reconstructs the image from the compressed representation."""
    compressed_data = np.dot(data, selected_eigenvectors)  # Project the data into the lower-dimensional space
    reconstructed_data = np.dot(compressed_data, selected_eigenvectors.T)  # Reconstruct the original data from the compressed representation
    reconstructed_data = reconstructed_data * std + mean  # Denormalize the data (revert to the original scale)
    reconstructed_data = np.clip(reconstructed_data, 0, 255).astype(np.uint8)  # Ensure pixel values are between 0 and 255, and convert to uint8 type
    return reconstructed_data  # Return the reconstructed image data

def experiment_pca(image_data, num_components):
    """Performs PCA compression and decompression for a given number of components."""
    rows, cols = image_data.shape  # Get the number of rows and columns of the image
    flattened_data = image_data.flatten().reshape(rows, cols)  # Flatten the image data to a 1D array and reshape
    mean = np.mean(flattened_data, axis=0)  # Calculate the mean of the image data
    std = np.std(flattened_data, axis=0)  # Calculate the standard deviation of the image data
    standardized_data = center_and_standardize(flattened_data)  # Standardize the data (center and scale)
    covariance_matrix = calculate_covariance_matrix(standardized_data)  # Calculate the covariance matrix of the standardized data
    eigenvalues, eigenvectors = eigen_decomposition(covariance_matrix)  # Perform eigen decomposition to get eigenvalues and eigenvectors
    sorted_indices = np.argsort(eigenvalues)[::-1]  # Sort the eigenvalues in descending order
    sorted_eigenvectors = eigenvectors[:, sorted_indices]  # Sort the eigenvectors according to the sorted eigenvalues
    selected_eigenvectors = sorted_eigenvectors[:, :num_components]  # Select the top 'num_components' eigenvectors (principal components)
    reconstructed_image = reconstruct_image(standardized_data, selected_eigenvectors, mean, std)  # Reconstruct the image using the selected principal components
    return reconstructed_image.reshape(rows, cols)  # Reshape the reconstructed image back to its original shape

# Experiment with different numbers of components
num_components_list = [10, 20, 50, 100, 150]  # List of different numbers of principal components to experiment with
reconstructed_images = {}  # Dictionary to store reconstructed images for each number of components

# Loop through each number of components in the list
for num_components in num_components_list:
    reconstructed_images[num_components] = experiment_pca(image_data, num_components)  # Perform PCA and get the reconstructed image

# Display the results
plt.figure(figsize=(12, 6))  # Set the figure size for the plot
plt.subplot(2, 3, 1)  # Create the first subplot for the original image
plt.imshow(image_data, cmap='gray')  # Display the original image in grayscale
plt.title('Original Image')  # Set the title for the original image
plt.axis('off')  # Remove the axis labels for the original image

# Loop through each reconstructed image and display it
for i, num_components in enumerate(num_components_list):
    plt.subplot(2, 3, i + 2)  # Create a subplot for each reconstructed image
    plt.imshow(reconstructed_images[num_components], cmap='gray')  # Display the reconstructed image in grayscale
    plt.title(f'{num_components} Components')  # Set the title with the number of components used
    plt.axis('off')  # Remove the axis labels for each reconstructed image

plt.tight_layout()  # Adjust the layout to ensure the subplots fit well
plt.show()  # Display the plot
```
